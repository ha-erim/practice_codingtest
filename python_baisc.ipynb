```python
# 배열 원소가 리스트일때 정렬
list_arr = [[1,2],[1,5],[3,8],[4,7]]
# x[0]을 기준으로 정렬. x[0]이 같다면 x[1]을 기준으로 오름차순
list_arr.sort(key=lambda x:(x[0],x[1]))
print(list_arr)

# 내림차순
_list_arr = [[1,2],[1,5],[3,8],[4,7]]
# x[0]을 기준으로 정렬. x[0]이 같다면 x[1]을 기준으로 오름차순
_list_arr.sort(key=lambda x:(-x[0],-x[1]))
print(_list_arr)
```

### 순열, 조합

```python
# 조합
# combinations의 첫 번째 인자에 배열을 넣고, 두 번째 인자에는 nCm 이라면 m에 해당하는 값을 넣기
from itertools import combinations
print(list(combinations([1,2,3,4],2)))
print(list(combinations([1,2,3,4],3)))
print(list(combinations([1,2,3,4],4)))
```

```python
# 순열
# permutations 사용

from itertools import permutations
N, M = map(int, input().split())
arr = [str(i+1) for i in range(N)]

for e in list(permutations(arr, M)):
    print(" ".join(e))
```

### 빈도 계산

```python
from collections import Counter

print(Counter(["hi", "hey", "hi", "hi", "hello", "hey"]))
print(Counter("hello world"))
```

```python
# counter를 dictionary처럼!
counter = Counter('hello world!')
print(counter)
print(counter["o"], counter["l"])
counter["l"] += 1
print(counter)
```

```python
# 특정 키 찾기
if "o" in counter:
    print("o in counter!")
```

```python
# 가장 흔한 데이터 찾기
from collections import Counter
print(Counter('hello world').most_common())
```

```python
# 가장 갯수가 많은 K개의 데이터도 찾을 수 있음
from collections import Counter
print(Counter('hello world').most_common(3))
```

### 힙

```python
# 최소힙
'''
     1 <--- Root 
   /   \ 
  3     5 
 / \   / 
4   8 7
'''
import heapq
heap = []
heapq.heappush(heap,3)
heapq.heappush(heap, 1)  
heapq.heappush(heap, 10)  
heapq.heappush(heap, 5)  
heapq.heappush(heap, 8)

print(heap)
```

```python
# 힙의 길이
len(heap)
print(heap)
# 힙 루트 제거
heap.pop()
print(heap)
```

```python
# 최대 힙
# 힙에 넣을 때 -를 붙여서 음수로 넣고, 최종적으로 출력할 떄 -1을 다시 곱
import heapq
_heap = []

heapq.heappush(_heap, -3)
heapq.heappush(_heap, -1)  
heapq.heappush(_heap, -10)  
heapq.heappush(_heap, -5)  
heapq.heappush(_heap, -8)

print(_heap)
root = heapq.heappop(_heap)
print(-root)
```

### 덱

```python
# LIFO, FIFO 처리가 가능한 데이터
from collections import deque
deq = deque() # 덱 초기화
deq = deque([i for i in range(1,5)])
print(deq)
```

```python
# 왼쪽에 값 추가
deq.appendleft(7)
print(deq)
# 오른쪽에 값 추가
deq.append(1)
print(deq)
# 왼쪽에서 값 제거
print(deq.popleft())
print(deq)
# 오른쪽에서 값 제거
print(deq.pop())
print(deq)
# 길이
print("len "+str(len(deq)))
```

```python
# deque 회전 (양수 인자로 들어간만큼 시계방향 회전, 음수인경우 반시계)
deq.rotate(3)
print(deq)
deq.rotate(-1)
print(deq)
```
