### 응용 1 - 기본순회

```python
# Graph Dictionary
# 기본형태에서 크게 변화 X, 추가 조건 제시
# 1. 문제에서 특정 "종료" 조건 제시
# 2. 문제에서 특정 "이동" 조건 제시

from collections import deque

def bfs(start_node, graph):
    queue = deque([start_node])
    visited = set([start_node])
    
    while queue:
        curr_node = queue.popleft()

        # 문제에서 제시하는 "종료" 조건
        if curr_node == 'T':
            return curr_node

        for next_node in graph[curr_node]:
            if next_node not in visited:
                # 문제에서 제시하는 "이동" 조건
                if next_node != 'X':
                    visited.add(next_node)
                    queue.append(next_node)
    return -1
```

```python
# Graph 2D Array
# 기본형태에서 크게 변화 X, 추가 조건 제시
# 1. 문제에서 특정 "종료" 조건 제시
# 2. 문제에서 특정 "이동" 조건 제시

from collections import deque

def bfs(start_row, start_col, graph):
    dx = [-1, 1, 0, 0]
    dy = [0, 0,-1, 1]
    rows, cols = len(graph), len(graph[0])

    queue = deque([(start_row, start_col)])
    visited = set([(start_row, start_col)])

    while queue:
        row, col = queue.popleft()

        # 문제에서 제시하는 "종료" 조건
        if row == 'T' and col == 'T':
            return row, col
        
        for i in range(4):
            next_row = row + dx[i]
            next_col = col + dy[i]

            if 0 <= next_row < rows and 0 <= next_col < cols:
                if (next_row, next_col) not in visited:
                    # 문제에서 제시하는 "이동" 조건
                    if graph[next_row][next_col] != 'X':
                        visited.add((next_row, next_col))
                        queue.append((next_row, next_col))

    return -1
```

### 응용 2 - 최단경로, 최단거리

최단경로

```python
# Graph Dictionary, 최단 경로
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 문제에서 특정 "종료" 조건 제시

from collections import deque

def bfs_shortest_path(start_node, end_node, graph):
    # 1. {노드: [경로]}
    queue = deque([start_node])
    visited = set([start_node])
    path = {start_node: [start_node]}

    while queue:
        curr_node = queue.popleft()

        # 2. "종료" 조건
        if curr_node == end_node:
            return path[curr_node]
        
        for next_node in graph[curr_node]:
            if next_node not in visited:
                visited.add(next_node)
                queue.append(next_node)
                # 1. {다음노드: [현재경로]+[다음경로]}
                path[next_node] = path[curr_node]+[next_node]

    return -1
```

```python
# Graph 2D Array, 최단 경로
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 문제에서 특정 "종료" 조건 제시
# 3. 문제에서 특정 "이동" 조건 제시

from collections import deque

def bfs_shortest_path_array(start_x, start_y, end_x, end_y, graph):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    n, m = len(graph), len(graph[0])

    # 1. (좌표 x, 좌표 y, [경로])
    visited = set([(start_x, start_y)])
    queue =deque([(start_x, start_y, [(start_x, start_y)])])

    while queue:
        x, y, path = queue.popleft()

        # 2. "종료" 조건
        if x == end_x and y == end_y:
            return path
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m:
                if (nx, ny) not in visited:
                    # 3. "이동" 조건
                    if graph[nx][ny] == 1:
                        # 1. (다음좌표 x, 다음좌표y, [현재경로]+[다음좌표])
                        visited.add((nx,ny))
                        queue.append((nx,ny,path+[(nx,ny)]))
                        
    return -1
```

최단거리

```python
# Graph Dictionary, 최단 거리
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 문제에서 특정 "종료" 조건 제시

from collections import deque

def bfs_shortest_distance(start_node, end_node, graph):
    # 1. {노드: 거리}
    queue = deque([start_node])
    visited = set([start_node])
    dist = {start_node:0}

    while queue:
        curr_node = queue.popleft()

        # 2, "종료" 조건
        if curr_node == end_node:
            return dist[curr_node]
        
        for next_node in graph[curr_node]:
            if next_node not in visited:
                visited.add(next_node)
                queue.append(next_node)
                # 1. {다음노드 : 현재거리 + 1}
                dist[next_node] = dist[curr_node] + 1
    
    return -1
```

```python
# Graph 2D Array, 최단 거리
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 문제에서 특정 "종료" 조건 제시
# 3. 문제에서 특정 "이동" 조건 제시

from collections import deque

def bfs_shortest_distance_array(start_x, start_y, end_x, end_y, graph):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    n, m = len(graph), len(graph[0])

    # 1. (좌표x, 좌표y, 거리)
    visited = set([(start_x, start_y)])
    queue = deque([(start_x, start_y, 1)])

    while queue:
        x, y, dist = queue.popleft()

        # 2. "종료" 조건
        if x==end_x and y==end_y:
            return dist
        
        for i in range(x):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m:
                if (nx, ny) not in visited:
                    # 3. "이동" 조건
                    if graph[nx][ny] == 1:
                        # 1. (다음좌표x, 다음좌표y, 현재거리+1)
                        visited.add((nx,ny))
                        queue.append((nx, ny, dist+1))
    return -1
```

최단경로, 최단거리 한 번에 구하기

```python
# Graph Dictionary, 최단 거리
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 문제에서 특정 "종료" 조건 제시
# 3. 문제에서 특정 "이동" 조건 제시

from collections import deque

def BFS_dict(start_node, end_node, graph):
    # 1. {노드: [경로]}
    queue = deque([start_node])
    visited = set([start_node])
    path = {start_node: [start_node]}

    while queue:
        curr_node = queue.popleft()

        # 2,3. 최단 경로의 길이를 통한 최단거리
        if curr_node == end_node:
            return path[curr_node], len(path[curr_node])-1
        
        for next_node in graph[curr_node]:
            if next_node not in visited:
                visited.add(next_node)
                queue.append(next_node)
                # 1. {다음노드:; [현재경로]+[다음경로]}
                path[next_node] = path[curr_node] + [next_node]
    return [], -1
```

```python
# Graph 2D Array, 최단 거리
# 1. 문제에서 특정 "결과" 값 요구시
# 2. 최단경로 + 최단 거리
# 3. 문제에서 특정 "종료" 조건 제시
# 4. 문제에서 특정 "이동" 조건 제시

def BFS_array(start_x, start_y, end_x, end_y, graph):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    n, m = len(graph), len(graph[0])

    # 1. (좌표 x, 좌표 y, [경로])
    visited = set([(start_x, start_y)])
    queue = deque([(start_x, start_y, [(start_x, start_y)])])

    while queue:
        x, y, path = queue.popleft()

        # 2,3. 최단 경로의 길이를 통한 최단거리
        if x == end_x and y == end_y:
            return path, len(path)-1
        
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m:
                if (nx,ny) not in visited:
                    # 4. 추가 이동 조건
                    if graph[nx][ny] == 1:
                        # 1. (다음좌표x, 다음좌표y, [현재경로]+[다음경로])
                        visited.add((nx, ny))
                        queue.append((nx, ny, path+[(nx,ny)]))
    return [], -1
```

### 연결요소

```python
# Graph Dictionary
# 특정 노드를 기준으로 연결된 모든 요소 찾기
# 1. 첫 노드는 방문하지 않은 것으로 처리
# 2. 특정 노드를 기준으로 연결된 모든 요소 탐색
# 3. 특정 노드를 기준으로 연결된 모든 요소를 결과값에 추가

def bfs_connected_components(start_node, graph):
    # 1. 첫 노드 방문 X
    visited = set()
    queue = deque([start_node])
    connected_components = []

    while queue:
        base_node = queue.popleft()

        # 1. 기준 노드
        if base_node not in visited:
            component = []
            visited.add(base_node)
            queue.append(base_node)

            # 2. 기준 노드에 연결된 요소 탐색
            while queue:
                curr_node = queue.popleft()
                component.append(curr_node)
                for next_node in graph[curr_node]:
                    if next_node not in visited:
                        visited.add(next_node)
                        queue.append(next_node)
            
            #3. 각각의 덩어리로 묶인 연결 요소
            connected_components.append(component)
    return connected_components
```

```python
# Graph 2D Array
# 특정 노드를 기준으로 연결된 모든 요소 찾기
# 1. visited를 bfs()가 아닌 solution에서 생성
# 2. 특정 노드를 기준으로 연결된 모든 요소 탐색
# 3. 문제에서 특정 "이동" 조건 제시
# 4. 모든 노드를 순회하며 특정 노드의 기준점
# 5. solution()에서도 이동조건 검사, bfs()결과를 최종 결과에 추가

from collections import deque

# 1. solution()에게 visited를 받아 처리
def bfs(start_x, start_y, graph, visited):
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    n, m = len(graph), len(graph[0])
    visited[start_x][start_y] = True

    component = []
    queue = deque([(start_x, start_y)])
    while queue:	
        x, y = queue.popleft()
        component.append((x, y))	# 2. 기준노드에 연결된 요소 탐색

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            # 3. 추가 이동 조건 graph[nx][ny]
            if 0 <= nx < n and 0 <= ny < m:
                if graph[nx][ny] and not visited[nx][ny]:
                    visited[nx][ny] = True
                    queue.append((nx, ny))
                    
    return component

def solution(graph):
	# 1. bfs()에게 넘겨줄 visited 생성
    n, m = len(graph), len(graph[0])
    visited = [[False] * m for _ in range(n)]

	# 4. 모든 노드를 순회하며, 각각의 모든 연결요소 탐색
    answer = []
    for i in range(n):
        for j in range(m):
        	# 5. 중복되지 않도록 solution()에서 이동 조건 검사
            if graph[i][j] and not visited[i][j]:
                answer.append(bfs(i, j, graph, visited))
                
    return answer
```

### 땅따먹기

일반적으로 2D Array로 제공

연결 요소의 연장선상의 문제. 

solution()에서 모두 순회하며 다른 집합 탐색

```python
# Graph Dictionary
# 1. 첫 노드 방문처리 X
# 2. extend()를 통해 for next_node in graph[curr_node]: 생략

from collections import deque

def bfs_connected_components(start, graph):
    # 1.
    visited = set()
    queue = deque([start])
    connected_components = []

    while queue:
        curr_node = queue.popleft()
        # 2. extend로 for next_node 생략
        if curr_node not in visited:
            visited.add(curr_node)
            queue.extend(graph[curr_node])
            connected_components.append(curr_node)
    
    return connected_components
```

```python
# Graph 2D Array
# 1. 연결 요소인지 확인할 기준색을 추출하고 비교
# 2. 연결요소인 경우 다른 연결 요소집합이랑 구분할 수 있도록 새로운 색으로 변경

from collections import deque

def bfs_flood_fill(sx, sy, graph, new_color):
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]

    # 1. 기준 색 추출
    queue = deque([(sx, sy)])
    visited = set([(sx, sy)])
    base_color = graph[sx][sy]

    while queue:
        # 2. 새로운 하나의 색 변경
        x, y = queue.popleft()
        graph[x][y] = new_color

        for i in range(4):
            nx = x+ dx[i]
            ny = y + dy[i]

            #1. 시작 색이랑 같은지 검사
            if 0 <= nx < len(graph) and 0 <= ny < len(graph[0]):
                if (nx, ny) not in visited and graph[nx][ny] == base_color:
                    visited.add((nx,ny))
                    queue.append((nx,ny))
    return graph
```
